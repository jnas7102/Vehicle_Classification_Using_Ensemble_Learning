```python
test_datagen= ImageDataGenerator(rescale=1./255)
```

```python
test_set= test_datagen.flow_from_directory(directory= './test',
                                               target_size=(224,224), # As we choose 64*64 for our convolution model
                                               batch_size=32,
                                               class_mode='categorical', # for 2 class binary
                                               shuffle = False)
```

```python
def get_models():
    # load json and create model
    json_file = open('./vehicle_classifier.json', 'r')
    loaded_model_json = json_file.read()
    json_file.close()
    model_1 = model_from_json(loaded_model_json)
    # load weights into new model
    model_1.load_weights("./vehicle_classifier.h5")
    print("Loaded model from disk")
    model_2 =  keras.models.load_model('C:/Users/deexi/Downloads/vehicle_classifier')
    models = list()
    models.append(('vgg_finetuned', model_1))
    models.append(('custom_cnn', model_2))
    return models
```

```python
# fit and evaluate each model
models = get_models()
# scores = evaluate_models(models, test_set)
```

```python
total_images = test_set.n  
steps = total_images//32 
x , y = [] , []
for i in range(steps):
    a , b = test_set.next()
    x.extend(a) 
    y.extend(b)
np.array(x), np.array(y)
```

```python
x= np.array(x)
y= np.array(y)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
# makes prediction according to given models and given weights
def predict(models, data, weights=None):
    if weights is None:
        # default weights provide voting equality
        weights = [1 / (len(models))] * len(models)
    pred = np.zeros((data.shape[0], 11))
    for i, model in enumerate(models):
        pred += model.predict(data) * weights[i]
    return pred
    
# returns accuracy for given predictions
def evaluate(preds, weights=None):
    if weights is None:
        weights = [1 / len(preds)] * len(preds)
    y_pred = np.zeros((y.shape[0], 11))
#    print(weights)
    for i, pred in enumerate(preds):
        y_pred += pred * weights[i]
    y_pred = np.argmax(y_pred, axis=1)
    y_true = np.argmax(y, axis=1)
    return f1_score(y_true, y_pred, average='macro')


# precalculated predictions of all models
preds = []
# evaluate every model as single
for i, model in enumerate(models):
    print (model)
    pred = predict([model[1]], x)
    preds.append(pred)
    score = evaluate([pred])
    print(f'model {i + 1}: f1 score = {score:.4f}')

# evaluate ensemble (with voting equality)
ensemble_score = evaluate(preds)
print(f'ensemble: f1 score = {ensemble_score:.4f}')
```

```python
best_score = 0
best_weights = None
no_improvements = 0
while no_improvements < 10000: #patience
    
    # generate normalized weights
    new_weights = np.random.uniform(size=(len(models), ))
    new_weights /= new_weights.sum()
    
    # get the score without predicting again
    new_score = evaluate(preds, new_weights)
    
    # check (and save)
    if new_score > best_score:
        no_improvements = 0
        best_score = new_score
        best_weights = new_weights
        print(f'improvement: {best_score:.4f}')
    else:
        no_improvements += 1

print(f'best weights are {best_weights}')
```
